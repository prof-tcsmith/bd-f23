# Demonstration of reading from MongoDB and writing to BSON

```python
import pymongo # pymongo is a python driver for MongoDB

client = pymongo.MongoClient('mongodb://localhost:27017/')

db = client.tweets 
collection = db.tweet_set01  
```

Search the tweet_set01 to find all retweets. Print the firts 10 to the screen.

```python
query = { "text": { "$regex": "^RT" } } # search for any name that starts with RT
docs = collection.find(query).limit(10) # limit the number of results to 10

for i, doc in enumerate(docs):
    print(doc)
    if 1 > 10:
        break
```

Here, we store the data from the previous seach in a BSON file (these are retweets). Note: You will most likely need to edit the path to where you would like the data stored. This code will create the file if it does not exist, and overwrite it if it does.

```python
import bson.json_util as bju

fin = open("./data/retweets.json", "w")
fin.write(bju.dumps(list(docs), indent=2))
fin.close()
```
